##PyScript Nedir?
PyScript, kullanıcıların HTML arayüzünü ve Pyodide, WASM ve modern web teknolojilerinin gücünü kullanarak tarayıcıda zengin Python uygulamaları oluşturmasına olanak tanıyan bir framework’tür. PyScript framework’ü, her deneyim seviyesindeki kullanıcılara, sayısız uygulama ile etkileyici, öğrenmesi kolay bir programlama diline erişim sağlar.

PyScript’in kendi sayfasında (https://pyscript.net/) kullanılan ifade: “PyScript yalnızca HTML'dir, Python kitaplıklarının zengin ve erişilebilir ekosistemi sayesinde yalnızca biraz (tamam, belki çok fazla) daha güçlüdür.” şeklindedir.

##PyScript’in Çalışması
PyScript, Pyodide'a bağlıdır . Bu nedenle, WebAssembly'ye dayalı tarayıcı ve Node.js için bir Python dağıtımıdır.

WebAssembly , modern web tarayıcılarında çalıştırılabilen yeni bir kod türüdür. Yerele yakın performansla çalışan ve C/C++, C# ve Rust gibi diller sağlayan kompakt bir ikili biçime sahip, düşük düzeyli, derleme benzeri bir dildir. Web üzerinde çalışabilmeleri için bir derlemedir.


##PyScript Temel Bileşenleri
*   **Tarayıcıda Python (Python in the browser):** Sunucu tarafı yapılandırmasına bağlı kalmadan açılan içeriği, harici dosya barındırmayı ve uygulama barındırmayı etkinleştirmeyi sağlar.
*   **Python Ekosistemi (Python Ecosystem):** Birçok popüler Python paketini ve bilimsel yığını çalıştırmayı sağlar. (numpy, pandas, scikit-learn ve daha fazlası gibi)
*   **Python ile JavaScript (Python with JavaScript):** Python ve Javascript nesne ve namespace’leri  arasında çift yönlü iletişim sağlar.
*   **Ortam Yönetimi (Environment Management):** Kullanıcıların, sayfa kodunun çalıştırılması için hangi paketlerin ve dosyaların dahil edileceğini tanımlamasına izin verir.
*   **Görsel Uygulama Geliştirme (Visual Application Development):** Button’lar, container’lar, textbox’ları ve daha fazlası gibi hazır olarak sunulan UI bileşenlerini kullanın
*   **Esnek Framework (Flexible Framework):** Plug ve extens edilebilen (Takılabilir ve genişletilebilir) Yeni bileşenleri doğrudan Python'da oluşturmak ve paylaşmak için kullanılabilecek esnek bir framework sunar.

##PyScript'in Dezavantajı
PyScript'in ana dezavantajı, destek eksikliğidir. PyScript, Pycon 2022 sırasında Python geliştiricilerine duyuruldu. Büyük ve heyecan verici bir olaydı, ancak topluluk beklentileri PyScript'in mevcut durumundan daha yüksekti. 

Zayıf topluluk desteği nedeniyle, gelecekte PyScript geliştirmesi daha yavaş olabilir. Python da son beş yıla kadar o kadar popüler değildi çünkü bilinmiyordu.

Python'un genel olarak JavaScript'in arama talebini nasıl bastırdığı Haziran 2022’ye ait grafikte görülüyor (Mavi Python, Kırmızı Javascript, Sarı PHP). Python'un popülaritesinin artmasının ana nedeni “pandas” kütüphanesidir.

Bu nedenle, ML ve Veri Bilimi Python'un ana odak noktalarıdır, ancak bu şekilde devam etmesi gerekmez. Bu nedenle, web geliştirmenin geleceği için PyScript ciddi şekilde dikkate alınmalıdır.



##PyScript’in Çalıştığı Mekanizma
Piyodid, PyScript'in temelidir. Piyodide aşina olmayan çok fazla veri bilimcisi yoktur. Yani WebAssembly'ye dayalı bir Python dağıtımıdır.

##WebAssembly ve WebAssembly Çalışması 
WebAssembly, Python programcılarının web sayfaları oluşturmasını sağlayan teknolojidir. Okunabilir format kullanır. “.wat” metin biçimi dili daha sonra tarayıcıların okuyabileceği bir ikili “.wasm” biçimine çevrilir. 
Bunun sayesinde geliştiriciler, herhangi bir dilde kod yazar, bir WebAssembly'de derlenir ve ardından bir web tarayıcısında çalıştırılır.
PyScript artık tarayıcıda Python kodu yazılmasına ve çalıştırılmasına izin veriyor. Gelecekte diğer dilleri de desteklemesi umulmaktadır. PyScript kullanılırken yalnızca Piyodid destekli kitaplıklar kullanılmaktadır.
